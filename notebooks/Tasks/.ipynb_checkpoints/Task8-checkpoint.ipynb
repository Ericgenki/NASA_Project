```python
import netCDF4 as nc
import numpy as np
from glob import glob
import xarray as xr
```

```python
file_path = glob('/home/hboi-ouri/Projects/RS_Files/Task8/NC Files/*_MODIS_L2.nc')
dset = nc.Dataset(file_path[0], "r")

dset.variables.keys()
```

```python

R443 = dset.variables["Rrs_443"][:]
R488 = dset.variables["Rrs_488"][:]
R531 = dset.variables["Rrs_531"][:]
R555 = dset.variables["Rrs_555"][:]
R667 = dset.variables["Rrs_667"][:]

```

```python
def get_CDOM300(R443, R488, R531, R555, R667, lambd=300):
        
        alpha = -.0206
        beta = -.6128
        gamma = -.007
        delta = -.4944
        epsilon = .9362
        zeta = .9666

        aCDOM300 = np.exp(alpha * np.log(R443) +
                      beta * np.log(R488) +
                      gamma * np.log(R531) +
                      delta * np.log(R555) +
                      epsilon * np.log(R667) + zeta)

        return aCDOM300
    
    
def get_CDOM355(R443, R488, R531, R555, R667, lambd=355):
                
        alpha = .0376
        beta = -.8714
        gamma = -.0352
        delta = -.2739
        epsilon = .9591
        zeta = -.1071

        aCDOM355 = np.exp(alpha * np.log(R443) +
                      beta * np.log(R488) +
                      gamma * np.log(R531) +
                      delta * np.log(R555) +
                      epsilon * np.log(R667) + zeta)

        return aCDOM355
    
    
def get_S275_295(R443, R488, R531, R555, R667):

        alpha = -.0537
        beta = .2689
        gamma = .1017
        delta = -.2097
        epsilon = -.0893
        zeta = -3.6853

        S275_295 = np.exp(alpha * np.log(R443) +
                      beta * np.log(R488) +
                      gamma * np.log(R531) +
                      delta * np.log(R555) +
                      epsilon * np.log(R667) + zeta)

        return S275_295


def get_DOC(R443, R488, R531, R555, R667):
                    
        alpha = -.0206
        beta = -.6128
        gamma = -.007
        delta = -.4944
        epsilon = .9362
        zeta = .9666

        old_aCDOM300 = np.exp(alpha * np.log(R443) +
                      beta * np.log(R488) +
                      gamma * np.log(R531) +
                      delta * np.log(R555) +
                      epsilon * np.log(R667) + zeta)
        
        alpha = -.0537
        beta = .2689
        gamma = .1017
        delta = -.2097
        epsilon = -.0893
        zeta = -3.6853

        S275_295 = np.exp(alpha * np.log(R443) +
                      beta * np.log(R488) +
                      gamma * np.log(R531) +
                      delta * np.log(R555) +
                      epsilon * np.log(R667) + zeta)
        
        DOC = (old_aCDOM300)/(np.exp(-15.05 - 33.95 * S275_295) + 
                            np.exp(-1.502 - 104.3 * S275_295))

        return DOC
    
```

```python

        
        
aCDOM300 = get_CDOM300(R443, R488, R531, R555, R667)
aCDOM355 = get_CDOM355(R443, R488, R531, R555, R667)
S275295 = get_S275_295(R443, R488, R531, R555, R667)
DOC = get_DOC(R443, R488, R531, R555, R667)
```
