```python
poi = ee.Geometry.Point([29.038136, -91.448196])
s2 = ee.ImageCollection('MODIS/061/MYD09A1').filterBounds(poi).filterDate('2019-11-20', '2019-11-30').min()

dem = ee.Image(s2)


def add_ee_layer(self, ee_image_object, vis_params, name):
  map_id_dict = ee.Image(ee_image_object).getMapId(vis_params)
  folium.raster_layers.TileLayer(
    tiles = map_id_dict['tile_fetcher'].url_format,
    attr = 'Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    name = name,
    overlay = True,
    control = True
  ).add_to(self)

# Add EE drawing method to folium.
folium.Map.add_ee_layer = add_ee_layer

# Set visualization parameters.
vis_params = {
  'min': 0,
  'max': 4000,
  }

# Create a folium map object.
my_map = folium.Map(location=[20, 0], zoom_start=3)

# Add the elevation model to the map object.
my_map.add_ee_layer(dem.updateMask(dem.gt(0)), vis_params, 'DEM')

# Add a layer control panel to the map.
my_map.add_child(folium.LayerControl())

# Display the map.
display(my_map)
```

```python
sat = ee.ImageCollection("MODIS/061/MYD09GA")

idate = '2018-04-05'

sat_img = sat.select('sur_refl_b03', 'sur_refl_b04').filterDate(idate).first()

sat_vis_params = {
    'min': -100.0,
    'max': 8000.0
}

lat, lon = -91, 29

my_map = folium.Map(location=[lat, lon], zoom_start=7)


def add_ee_layer(self, ee_image_object, vis_params, name):
    """Adds a method for displaying Earth Engine image tiles to folium map."""
    map_id_dict = ee.Image(ee_image_object).getMapId(vis_params)
    folium.raster_layers.TileLayer(
        tiles=map_id_dict['tile_fetcher'].url_format,
        attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
        name=name,
        overlay=True,
        control=True
    ).add_to(self)

folium.Map.add_ee_layer = add_ee_layer


my_map.add_ee_layer(sat_img, sat_vis_params, 'Band 03')

my_map.add_child(folium.LayerControl())

display(my_map)


```
